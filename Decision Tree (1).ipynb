```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
```

```python
Credit_score_data = pd.read_csv('cleaned_data.csv')
```

```python
Credit_score_data.head()
```

```python
Credit_score_data.shape
```

```python
x = Credit_score_data.drop('Credit_Score',axis = 1)
```

```python
x.head()
```

```python
y = Credit_score_data['Credit_Score']
```

```python
y
```

```python
label_encoders = {}
```

```python
for column in x.select_dtypes(include=['object']).columns:
    le = LabelEncoder()
    x[column] = le.fit_transform(x[column])
    label_encoders[column] = le  # Store label encoders in case you need them later
```

```python
X_train, X_test, Y_train,Y_test = train_test_split(x, y, test_size = 0.3, random_state = 42)
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier(random_state = 42)
```

```python
model.fit(X_train, Y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
Y_test
```

```python
from sklearn import metrics
```

```python
print(metrics.accuracy_score(Y_test, y_pred))
```

```python
rf_classifier = RandomForestClassifier(random_state=42)
rf_classifier.fit(X_train, Y_train)
rf_y_pred = rf_classifier.predict(X_test)
```

```python
print("Decision Tree Classifier Report:")
print(classification_report(Y_test, y_pred))
```
